```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style='whitegrid')
```

```python
df=pd.read_csv('happy.csv')
df.head()
```

```python
print("Shape:", df.shape)
df.info()
df.describe()
```

```python
#Missing Values Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap='Reds')
plt.title("Missing Value Heatmap")
plt.show()
df.isnull().sum()
```

```python
#Top 10 Happiest Countries
top10 = df.sort_values('Score', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x='Score',y='Country or region',data=top10, palette='viridis')
plt.title('Top 10 Happiest Countries')
plt.xlabel('Happiness Score')
plt.ylabel('Country')
plt.show()
```

```python
#Happiness Score by Region
plt.figure(figsize=(12,6))
sns.boxplot(x='Country or region', y='Score', data=df)
plt.xticks(rotation=45)
plt.title('Happiness Score Distribution by Region')
plt.show()
```

```python
#Correlation Matrix
numerics = df.select_dtypes(include=[np.number])
plt.figure(figsize=(10,8))
sns.heatmap(numerics.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
#Pairplot for Key Variables
sns.pairplot(df[['Overall rank',	'Country or region',	'Score',	'GDP per capita',	'Social support',	'Healthy life expectancy',	'Freedom to make life choices',	'Generosity',	'Perceptions of corruption']], corner=True)
plt.suptitle('Pairwise Plot of Key Factors', y=1.02)
plt.show()
```

```python
#Effect of Economy on Happiness
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='GDP per capita', y='Score')
plt.title('Economy vs. Happiness Score')
plt.show()
```

```python
#Effect of Trust in Government
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Trust (Government)', y='Score', hue='Region')
plt.title('Trust in Government vs. Happiness Score')
plt.show()
```

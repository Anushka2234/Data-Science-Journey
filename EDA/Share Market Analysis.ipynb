```python
df=pd.read_csv('share_market.csv')
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Date'],df['Close'], label = 'Close Price', color = 'blue')
plt.title('Stock Price Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
sns.histplot(df['Close'], bins=30, kde=True, color='skyblue')
plt.title('Histogram of Close Prices')
plt.xlabel('Close Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['Close'], color='lightcoral')
plt.title('Boxplot of Close Prices')
plt.xlabel('Close Price')
plt.tight_layout
```

```python
df['PriceChange'] = df['Close']-df['Open']
df['PctChange']=df['Close'].pct_change()*100
plt.figure(figsize=(12,6))
plt.plot(df['Date'],df['PctChange'], color='purple' ,label='Daily % Change')
plt.title('Daily Percentage Change in Close Prices')
plt.xlabel('Date')
plt.ylabel('Percentage Change(%)')
plt.axhline(0, color='gray', linestyle='--',linewidth=1)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Volume', y='Close', alpha=0.6)
plt.title('Scatter Plot: Volume vs Close Price')
plt.xlabel('Volume')
plt.ylabel('Close Price')
plt.tight_layout()
plt.show()
```

```python
#Linear regression dataset
import math
import numpy as np
import matplotlib.pyplot as plt

data = np.array([
    [0.05, 0.12],
    [0.18, 0.22],
    [0.31, 0.35],
    [0.42, 0.38],
    [0.5, 0.49]])
print(data)
```

```python
#split into inputs and outputs
x, y = data[:,0], data[:,1]
x = x.reshape((len(x), 1))
plt.scatter(x, y)
plt.show()
```

```python
#Calculate  cofficients
b = np.linalg.pinv(x).dot(y)
print(b)

#predict using coefficient
yhat = x.dot(b)


plt.scatter(x,y)
plt.plot(x, yhat, color='red')
plt.show()
```

```python
#equation in presence of noise
n = 100
x = np.linspace(-10,10,n)
a = 2
b = 3
y = a*x+b
mean = 0
var = 1
sd = 5

#plotting the graph
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
plt.plot(x,y,'black')

y = a*x+b+sd*np.random.normal(mean, var, n)

plt.scatter(x,y,color='red')
plt.show()
```

```python
ones = np.ones(n)
A = np.array([x,ones])
Ainv = np.linalg.pinv(A.T)
p = np.dot(Ainv,y)
print("Output of Pseudo-Inverse {}".format(p))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

yhat = A.T.dot(p)
r2 = r2_score(y, yhat)
print("R^2 value is: {}".format(r2))

MAE = mean_absolute_error(y, yhat)
print("MAE value is: {}".format(MAE))
```

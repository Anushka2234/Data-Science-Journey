```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Movie_Data.csv")
```

```python
df
```

```python
#Movie with highest Box office earning
highest_earning_movie = df.loc[df['box_office'].idxmax()]
print(highest_earning_movie[['title','genre','rating','release_year']].to_string())
```

```python
# What is the average rating of all movies in the dataset?
average_rating = df['rating'].mean()
print(average_rating)
```

```python
# Find the median release year
median_release_year = df['release_year'].median()
print(int(median_release_year))
```

```python
# Find the mode of movie genres
mode_genre = df['genre'].mode()[0]
print(mode_genre)
```

```python
# Filter the dataset to show only movies with a rating greater than 7
high_rated_movies = df[df['rating']>7]
print(high_rated_movies.to_string())
```

```python
# Sort the movies by box office earnings in descending order and display the top 10 movies
top_10_movies = df.sort_values(by='box_office', ascending=False).head(10)
print(top_10_movies)
```

```python
# Create a new column called "profitability"
df['profitability']=df['box_office']/100000000
print(df[['title','box_office','profitability']].to_string())
```

```python
# Group the dataset by genre and find the average rating and average box office earnings for each genre
genre_analysis = df.groupby('genre').agg(
    average_rating=('rating','mean'),
    average_box_office=('box_office','mean')).reset_index()
print(genre_analysis.to_string())
```

```python
# Plot a line graph to show the trend of the average rating over the years.
# Without matplotlib, we can represent the trend data as a table or list.
average_rating_by_year = df.groupby('release_year')['rating'].mean().reset_index()
average_rating_by_year = average_rating_by_year.sort_values(by='release_year')

print("\nTrend of Average Movie Rating Over the Years:")
print(average_rating_by_year.to_string(index=False))
```

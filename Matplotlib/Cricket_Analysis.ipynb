```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cricket_players.csv')
df
```

```python
df.rename(columns = {'Runs':'Runs Scored','Wickets':'Wickets Taken','Average':'Batting Average','Hundreds':'Hundreds Scored','Fifties':'Fifties Scored'}, inplace=True)
print(df.head(5))
```

```python
# 1. Bar plot: Total runs scored by players from different countries
plt.figure(figsize=(10,6))
country_runs = df.groupby('Country')['Runs Scored'].sum().reset_index()
sns.barplot(x='Country', y='Runs Scored', data=country_runs,hue='Country',palette='viridis', legend=False)
plt.title('Total Runs Scored by Country')
plt.xlabel('Country')
plt.ylabel('Total Runs Scored')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 2. Box plot: Distribution of Batting Average across different countries
plt.figure(figsize=(10,6))
sns.boxplot(x='Country',y='Batting Average',data=df,palette='pastel')
plt.title('Distribution of Batting Average by Country')
plt.xlabel('Country')
plt.ylabel('Batting Average')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Select numerical columns for correlation
numerical_cols = ['Matches Played', 'Runs Scored', 'Wickets Taken',
                  'Batting Average', 'Strike Rate', 'Hundreds Scored', 'Fifties Scored']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Cricket Player Statistics')
plt.show()
```

```python
# Modified Bar plot: Total runs scored by players from different countries
plt.figure(figsize=(10, 6))
# Fix for FutureWarning: Pass 'Country' to hue and set legend=False
sns.barplot(x='Country', y='Runs Scored', data=country_runs, hue='Country', palette='Set2', legend=False) # Changed palette
plt.title('Total Runs Scored by Country (Customized Appearance)') # Adjusted title
plt.xlabel('Player\'s Country') # Adjusted x-axis label
plt.ylabel('Aggregated Total Runs Scored') # Adjusted y-axis label
plt.grid(True, linestyle=':', alpha=0.6) # Added gridlines with different style
plt.show()
```

```python
# Scatter plot: Relationship between Matches Played and Runs
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Matches Played', y='Runs Scored', data=df, hue='Country', s=100, alpha=0.8, style='Country') # Added hue and style for better insight
plt.title('Relationship between Matches Played and Runs Scored by Player') # Added title
plt.xlabel('Number of Matches Played') # Added x-axis label
plt.ylabel('Total Runs Scored') # Added y-axis label
plt.grid(True, linestyle='--', alpha=0.5) # Added gridlines
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left') # Move legend outside
plt.tight_layout() # Adjust layout to prevent legend cut-off
plt.show()
```

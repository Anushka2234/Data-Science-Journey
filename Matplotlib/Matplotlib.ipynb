```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[14,15,6,7,8]
plt.plot(x,y)
plt.title("This is my first graph")
plt.xlabel("Price of the Sugar")
plt.ylabel("Qty of the Sugar")
plt.show() 
```

```python
#Line Plot
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
temperature = [22, 21, 23, 24, 22]

plt.plot(days, temperature,'v-b')
plt.title("Temperature over days")
plt.xlabel("day")
plt.ylabel("temperature (C)")
plt.grid(True)
plt.show()

```

```python
days = [1,2,3,4,5,6,7]
gold_prices = [8975,9923,8765,9115,8931,8769,8721]
plt.plot(days, gold_prices, marker='o',linestyle='-',color='gold',label='Gold Price')
plt.xlabel('Days')
plt.ylabel('Price(USD)')
plt.title('Gold Market Trend')
plt.legend()
plt.show()
```

```python
epochs = [1,2,3,4,5]
training_accuracy = [0.6, 0.72,0.80,0.85,0.88]
validation_accuracy = [0.58,0.70,0.78,0.82,0.84]
plt.plot(epochs, training_accuracy, marker='o', label='Training Accuracy')
plt.plot(epochs, validation_accuracy, marker='s', label='Validation Accuracy')
plt.title('Model Accuracy per Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Bar chart
category=['Cars','Bikes','Trucks']
values=[78,67,45]
plt.bar(category,values,color=['r','g','b'])
plt.xlabel("Vehicle category")
plt.ylabel("Sales")
plt.show()
           
```

```python
products=['Product A','Product B','Product C','Product D']
sales=[20,40,35,50]
plt.style.use('Solarize_Light2')
plt.bar(products, sales, color='skyblue')
plt.title("Sales by product")
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
cat = ['Car', 'Bike', 'Truck']
val = [78, 67, 45]

plt.bar(cat, val, color=['skyblue', 'orange', 'lightgreen'])
plt.style.use('ggplot')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.title('Number of Vehicles by Type')
plt.grid(axis='y', linestyle='--', alpha=0.8)
plt.show()
```

```python
#Histogram
import numpy as np 
data =np.random.randn(1000)
plt.hist(data,bins=13, edgecolor='black')
plt.title('Histogram')
plt.show()
```

```python
data = np.random.randn(1000)

plt.figure(figsize=(2, 3))
plt.hist(data, bins=12, edgecolor="black", color="skyblue", alpha=0.7)
plt.legend(["Random Data"], loc="upper right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.title("Histogram of Random Data", fontsize=14, fontweight="bold")
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency",fontsize=12)
plt.show()
```

```python
x=np.random.rand(59)
y=np.random.randn(59)
plt.scatter(x,y)
```

```python
#Pie Chart
departments = ['Cardiology','Neurology','Pediatrics','Orthopedics']
visits = [120,90,150,100]
plt.pie(visits,labels=departments, autopct='%1.1f%%',startangle=140)
plt.title('Department wise')
plt.axis('equal')
plt.show()
```

```python
departments = ['Cardiology','Neurology','Pediatrics','Orthopedics']
visits = [120,90,150,100]
plt.pie(visits,labels=departments, autopct='%1.1f%%',startangle=140,shadow=True)
plt.title('Department wise')
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import io 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('product_sales.csv')
```

```python
df
```

```python
#part1 - plot line plot(Sales Trend of Product A)
plt.figure(figsize=(10,6))
plt.plot(df['Date'],df['Product A'],marker='o', linestyle='-',color='skyblue')
plt.title('Sales Trend of Product A Over Time')
plt.xlabel('Date')
plt.ylabel('Product  A Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Multi-Line Plot (Sales Trends of Product A, B, C)
plt.figure(figsize=(12, 7))
plt.plot(df['Date'], df['Product A'],label='Product A', marker='o', linestyle='-')
plt.plot(df['Date'], df['Product B'],label='Product B', marker='s', linestyle='--')
plt.plot(df['Date'], df['Product C'], label='Product C', marker='^', linestyle=':')
plt.title('Sales Trend of Product A, Product B, and Product C Over time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Bar Chart (Total Year sales)
total_sales = df[['Product A', 'Product B', 'Product C']].sum()
highest_selling_product = total_sales.idxmax()
colors = ['skyblue' if product!=highest_selling_product else 'lightcoral' for product in total_sales.index]
plt.figure(figsize=(8,6))
plt.bar(total_sales.index, total_sales.values, color=colors)
plt.title('Total Yearly Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.grid(axis='y')
plt.show()
```

```python
# Part 4: Histogram (Daily Demand Fluctuations)
# Simulate daily demand fluctuations
np.random.seed(42)
demand_fluctuations = np.random.randn(1000)
plt.figure(figsize=(10,6))
plt.hist(demand_fluctuations, bins=30, edgecolor='black', alpha=0.7, color='purple')
plt.title('Distribution of Daily Demand Fluctuations(Simulated)')
plt.xlabel('Demand Fluctuation')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
# Part 5 (Bonus): Apply Plot Styles
# Style 1: dark_background
plt.style.use('dark_background')
plt.figure(figsize=(10,6))
plt.plot(df['Date'],df['Product A'], marker='o', linestyle='-',color='cyan')
plt.title('Sales Trend of Product A (dark_background style)')
plt.xlabel('Date')
plt.ylabel('Product A Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Style2: ggplot
plt.style.use('ggplot')
plt.figure(figsize=(10,6))
plt.plot(df['Date'],df['Product B'], marker='s',linestyle='--', color='orange')
plt.title('Sales Trend of Product B (ggplot style)')
plt.xlabel('Date')
plt.ylabel('Product B Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

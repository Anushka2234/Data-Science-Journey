```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import ipywidgets as widgets
```

```python
#loading Images
img = cv2.imread('Animal.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
#Defining adjustment function
def adjust_brightness_contrast(image, brightness=0, contrast=0):
    beta = brightness
    alpha = (contrast+100)/100
    adjusted = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
    return adjusted
```

```python
adjusted_img = adjust_brightness_contrast(img, brightness=50, contrast=30)

# Convert to RGB for display
adjusted_rgb = cv2.cvtColor(adjusted_img, cv2.COLOR_BGR2RGB)

# Show result
plt.imshow(adjusted_rgb)
plt.title("Brightness +50, Contrast +30")
plt.axis("off")
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(img_rgb)
axs[0].set_title("Original")
axs[0].axis('off')

axs[1].imshow(adjusted_rgb)
axs[1].set_title("Adjusted")
axs[1].axis('off')

plt.tight_layout()
plt.show()
```

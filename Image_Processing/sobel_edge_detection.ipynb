```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading Image in Grayscale
img = cv2.imread('Animal.jpg')
plt.imshow(img, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis('off')
plt.show()
```

```python
#Apply sobel filter
#computing gradient in x and y direction
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

#calculate gradient magnitude
sobel_magnitude = np.sqrt(sobel_x**2+sobel_y**2)
sobel_magnitude = np.uint8(np.clip(sobel_magnitude, 0, 255))

```

```python
#sobel outputs
plt.figure(figsize=(12,4))

plt.subplot(1, 3, 1)
plt.imshow(np.absolute(sobel_x), cmap='gray')
plt.title('Sobel X')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(np.absolute(sobel_y), cmap='gray')
plt.title('Sobel Y')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(sobel_magnitude, cmap='gray')
plt.title('Edge Magnitude')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
#Threshold to show strong edges only
_, threshold = cv2.threshold(sobel_magnitude, 100, 255, cv2.THRESH_BINARY)
plt.imshow(threshold, cmap='gray')
plt.title("Threshold Edges")
plt.axis('off')
plt.show()
```
